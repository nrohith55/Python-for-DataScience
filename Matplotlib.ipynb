# Matplotlib

## Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
Benifits of Matplotlib library are as follows

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
plt.scatter(x,y,c='g')# Plotting scatter diagram
plt.xlabel("X_axis")
plt.ylabel("Y_axis")
plt.title("2D Diagram")
plt.show()
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
#Creating subplots

plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,'ro')
plt.subplot(2,2,3)
plt.plot(x,y,'g*')
plt.subplot(2,2,4)
plt.plot(x,y,'b--')
```

```python
np.pi # Pi value
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Sine wave")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Cosine wave")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1) # To plot sine and cosine waves togethef
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,"bo")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Sine wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos,"g*")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Cosine wave")

plt.show()
```

# Bar Plot

```python
x=[1,2,3]
y=[3,2,1]
plt.bar(x,y)
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
plt.show()
```

# Histogram

```python
a= np.array([10,20,30,45,6,70,80,89,88,72,65,43,45,6,7,100,200,500])
plt.hist(a)
plt.title("histogram") 
plt.show()
```

# Box plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False); 
```

# Pie chart

```python
labels = 'A', 'B', 'C', 'D'
sizes = [20, 10, 25, 45]
colors = ['g','r','b','grey']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot (L(Labels),S(Size),C(Colors),E(Explode))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
